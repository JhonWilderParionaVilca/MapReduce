Ecribimos nuestras funciones mapp y reduce

```python
%%writefile mapper.py
#!/usr/bin/env python
from __future__ import print_function
import sys

for line in sys.stdin:
    line = line.strip()
    words = line.split("=")
    for word in words:
        print(word,'=',1)
```

```python
%%writefile reducer.py
#!/usr/bin/env python
from __future__ import print_function

import sys


current_word = None
current_count = 0
word = None

print("----------------------------Contador de aprobados------------------------------")

for line in sys.stdin:
    line = line.strip()
    word, count = line.split('=', 1)
    try:
        count = int(count)
    except ValueError:
        continue
    if current_word == word:
        current_count += count
    else:
        if current_word:
            print(current_word,'=', current_count)
        current_count = count
        current_word = word
if current_word == word:
    print(current_word,'=', current_count)
            
```
